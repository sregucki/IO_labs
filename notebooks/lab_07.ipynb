## Zadanie 1


```python
# read text from file
with open("data/news.txt", "r") as f:
    text = f.read()
```

```python
# download all nltk data
import nltk

nltk.download("all")
```

```python
# tokenize text
from nltk.tokenize import word_tokenize

text_tokenized = word_tokenize(text)
print(text_tokenized)
print(f"Number of words: {len(text_tokenized)}")
```

```python
# remove stopwords
from nltk.corpus import stopwords

text_tokenized_no_stopwords = [
    word for word in text_tokenized if word.lower() not in stopwords.words("english")
]
print(text_tokenized_no_stopwords)
print(f"Number of words: {len(text_tokenized_no_stopwords)}")
```

```python
# lemmentize words
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

text_tokenized_lemmatized = [
    lemmatizer.lemmatize(word) for word in text_tokenized_no_stopwords
]
print(text_tokenized_lemmatized)
print(f"Number of words: {len(text_tokenized_lemmatized)}")
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# remove special characters
text_tokenized_lemmatized_no_punctuation = [
    word for word in text_tokenized_lemmatized if word.isalnum()
]

# count words
word_counts = Counter(text_tokenized_lemmatized_no_punctuation)
print(word_counts)

top_10_words = dict(word_counts.most_common(10))

# top 10 words
plt.bar(top_10_words.keys(), top_10_words.values())
plt.xticks(rotation=45)
plt.show()
```

```python
# word cloud
from wordcloud import WordCloud

wordcloud = WordCloud().generate_from_frequencies(word_counts)
plt.imshow(wordcloud, interpolation="bilinear")
```
